```python
import pandas as pd
import plotly.graph_objects as go

data.head()

fig = go.Figure(data = go.Scattergeo(
    lon = data["Longitude"],
    lat = data["Latitude"],
    text = data["Time"],
    mode = "markers",
    marker = dict(
        size = 12,
        opacity = 0.8,
        reversescale = True,
        autocolorscale = True,
        symbol = 'circle',
        line = dict(
            width = 1,
            color = 'rgba(102, 102, 102)'
        ),
        cmin = 0,
        color = data['Magnitude'],
        cmax = data['Magnitude'].max(),
        colorbar_title = "Magnitude"
    )
))

fig.update_layout(
    title = "Earthquakes in Turkey",
    geo = dict(
        scope = "world",
        showland = True,
    )
)

fig.write_html('first_figure.html', auto_open=True)
```
